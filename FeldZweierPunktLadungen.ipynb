Wir berechnen das elektrische Feld zweier Ladungen.  
Einmal wenn die Ladungen gleichnamig sind, und  
zum anderen mal, wenn die Ladungen ungleichnamig  
sind.  
  
Die Felder sollten den Maxwell'schen Gleichungen  
gehorchen.


```python
import matplotlib.pyplot as plt
import numpy as np

X, Y = np.meshgrid(np.arange(-20.5, 20.5, 1), np.arange(-20.5, 20.5, 1))
X1 = X + 10
X2 = X - 10

U1 = X1 / pow((X1**2 + Y**2), 3/2)
V1 = Y / pow((X1**2 + Y**2), 3/2)
U2 = X2 / pow((X2**2 + Y**2), 3/2)
V2 = Y / pow((X2**2 + Y**2), 3/2)

Up = U1 + U2
Vp = V1 + V2
```

FÃ¼r gleichnamige Ladungen:


```python
fig, ax = plt.subplots()
q = ax.quiver(X, Y, Up, Vp)

plt.show()
```

```python
fig, ax = plt.subplots()
q = ax.streamplot(X, Y, Up, Vp)

plt.show()
```

FÃ¼r ungleichnamige Ladungen:


```python
Un = U1 - U2
Vn = V1 - V2
```

```python
fig, ax = plt.subplots()
q = ax.quiver(X, Y, Un, Vn)

plt.show()
```

```python
fig, ax = plt.subplots()
q = ax.streamplot(X, Y, Un, Vn)

plt.show()
```
